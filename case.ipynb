```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings

plt.rc('font', family = 'AppleGothic')
filterwarnings('ignore')

df = pd.read_csv("dataset/sac_ver02.csv").query("`place`=='콘서트홀'")

df_pivot_all_raw = df.pivot_table(index=['seat_floor','seat_col'], columns=['seat_block','seat_num'], aggfunc='size', fill_value=0)
df_pivot_all_raw
```

```python
fc_list = df_pivot_all_raw.index.tolist() #seat_floor, seat_col
fc_list[0][0] #n층 n열

bn_list = df_pivot_all_raw.columns.tolist() #seat_block, seat_num
bn_list[0][0] #n블록 n좌석

new_list = [(x[0], y[0], x[1], y[1]) for x in fc_list for y in bn_list]
new_list

dfnew = df[:0].copy()

for i,(f,b,c,n) in enumerate(new_list):
    dfnew.loc[i,:] = -1
    dfnew.loc[i,'seat_floor'] = f
    dfnew.loc[i,'seat_block'] = b
    dfnew.loc[i,'seat_col'] = c
    dfnew.loc[i,'seat_num'] = n

dfcase = pd.concat([df,dfnew])
dfcase
```

```python
print(len(df))
print(len(dfnew))
print(len(dfcase))

dfnew.to_csv('df_case.csv',index=False)
```
