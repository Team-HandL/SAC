## 결측치 대치 가설에 대한 독립성검정 수행

```python
y = df.query('`member_yn` == 1').groupby(['membership_paid', 'age']).count()[['gender']].reset_index()
cross_table = y.pivot_table(index = 'membership_paid', columns = 'age', values = 'gender')
```

```python
chi_res = chi2_contingency(cross_table)
n = np.sum(cross_table.values)
t = np.min([len(cross_table.index)-1, len(cross_table.columns)-1])
print(f'stataistic chi2 : {chi_res[0]}', '   ',f'p-value : {chi_res[1]}')
print('sample number :', int(n))
print('phi coefficient :', np.sqrt(chi_res[0]/n),'   ', 'max phi coefficient :',  np.min([np.sqrt(len(cross_table.index)-1), np.sqrt(len(cross_table.columns)-1)]))
print('Cramer\'s V :', np.sqrt(chi_res[0]/(n*t)))
```

```python
y2 = df.query('`member_yn` == 1').groupby(['membership_free', 'age']).count()[['gender']].reset_index()
cross_table = y2.pivot_table(index = 'membership_free', columns = 'age', values = 'gender')
cross_table =cross_table.fillna(0)

chi_res = chi2_contingency(cross_table)
n = np.sum(cross_table.values)
t = np.min([len(cross_table.index)-1, len(cross_table.columns)-1])

print(f'stataistic chi2 : {chi_res[0]}', '   ',f'p-value : {chi_res[1]}')
print('sample number :', int(n))
print('phi coefficient :', np.sqrt(chi_res[0]/n),'   ', 'max phi coefficient :', np.min([np.sqrt(len(cross_table.index)-1), np.sqrt(len(cross_table.columns)-1)]))
print('Cramer\'s V :', np.sqrt(chi_res[0]/(n*t)))
```
